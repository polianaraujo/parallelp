```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Carregar dados do CSV
df = pd.read_csv("results_escalabilidade_forte.csv")

# 2. Agrupar para obter tempo mdio por nmero de threads
df_avg = df.groupby("Threads")["Time"].mean().reset_index()

# 3. Calcular speedup em relao ao tempo com 1 thread
tempo_serial = df_avg[df_avg["Threads"] == 1]["Time"].values[0]
df_avg["Speedup"] = tempo_serial / df_avg["Time"]

# 4. Plotar grfico de Speedup vs Threads
plt.figure(figsize=(10, 6))
plt.plot(df_avg["Threads"], df_avg["Speedup"], marker='o', linestyle='-', color='teal', label='Speedup real')
plt.plot(df_avg["Threads"], df_avg["Threads"], linestyle='--', color='gray', label='Speedup ideal (linear)')

plt.title("Escalabilidade Forte - Grfico de Speedup vs Nmero de Threads")
plt.xlabel("Nmero de Threads")
plt.ylabel("Speedup")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
