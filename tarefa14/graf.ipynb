```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# Lê o CSV
df = pd.read_csv("tempo_pingpong.csv")

# Criar diretório para salvar o gráfico
output_dir = "images"
os.makedirs(output_dir, exist_ok=True)

# Converte tamanhos para kB para facilitar visualização log
df["Tamanho (kB)"] = df["Tamanho(Bytes)"] / 1024

# Gráfico 1: Tempo médio vs Tamanho da mensagem
plt.figure(figsize=(10, 6))
plt.plot(df["Tamanho (kB)"], df["Tempo_Medio(s)"] * 1e6, marker='o', linestyle='-')
plt.xscale("log")
plt.xlabel("Tamanho da mensagem (kB, escala log)")
plt.ylabel("Tempo médio por troca (μs)")
plt.title("Tempo médio de troca MPI vs Tamanho da Mensagem")
plt.grid(True, which="both", ls="--")
plt.tight_layout()
plt.savefig("images/tempo_medio_vs_tamanho.png", dpi=300)
plt.show()

# Gráfico 2: Tempo total vs Tamanho da mensagem
plt.figure(figsize=(10, 6))
plt.plot(df["Tamanho (kB)"], df["Tempo_Total(s)"], marker='s', color='red')
plt.xscale("log")
plt.xlabel("Tamanho da mensagem (kB, escala log)")
plt.ylabel("Tempo total (s)")
plt.title("Tempo total de trocas MPI vs Tamanho da Mensagem")
plt.grid(True, which="both", ls="--")
plt.tight_layout()
plt.savefig("images/tempo_total_vs_tamanho.png", dpi=300)
plt.show()
```
