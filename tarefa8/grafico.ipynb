```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leitura do CSV
df = pd.read_csv("tempos_pi.csv")

# Grfico
plt.figure(figsize=(10, 6))
bars = plt.bar(df["Verso"], df["Tempo (s)"], color=["#FF6666", "#66B2FF", "#FF9966", "#66FF99"])
plt.title(" Tempo de Execuo das Verses da Estimativa de ", fontsize=14)
plt.ylabel("Tempo (segundos)", fontsize=12)
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.02, f"{yval:.3f}s", ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
